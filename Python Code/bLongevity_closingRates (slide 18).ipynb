```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tableau_blue = (31/255, 119/255, 180/255)
```

```python
df_long = pd.read_csv('/content/rft_bLongevity_ClosingSeason.csv', sep = ',')
df_long.head()
```

```python
# List of desired categories
desired_categories = ['Event Planning & Services', 'Hotels & Travel', 'Automotive', 'Food', 'Health & Medical']

# Filter rows where CATEGORY_NAME is in the list of desired categories
selected_df = df_long[df_long['CATEGORY_NAME'].isin(desired_categories)]
```

```python
average_longevity = selected_df.groupby('CATEGORY_NAME')['LONGEVITY_INDAYS'].mean()
```

```python
average_longevity.plot(kind='bar', rot=45, color=tableau_blue)
plt.ylim(1000, 3000)

# Add data labels on top of the bars
for i, value in enumerate(average_longevity):
    plt.text(i, value + 1, '{:.0f}'.format(value), ha='center', va='bottom')

plt.xlabel('Category Name')
plt.ylabel('Average Longevity (in days)')
plt.title('Average Longevity for Each Category')
plt.show()
```

```python
average_rft = selected_df.groupby('CATEGORY_NAME')['AVG_RETAIL_FOOT_TRAFFIC'].mean()
```

```python
average_rft.plot(kind='bar', rot=45, color=tableau_blue)
plt.ylim(6, 9)  # Adjust the values based on your desired range
for i, value in enumerate(average_rft):
    plt.text(i, value + 0.1, '{:.1f}'.format(value), ha='center', va='bottom')  # Adjust the format and offset as needed

plt.xlabel('Category Name')
plt.ylabel('Average RFT')
plt.title('Average RFT for Each Category')
plt.show()
```
